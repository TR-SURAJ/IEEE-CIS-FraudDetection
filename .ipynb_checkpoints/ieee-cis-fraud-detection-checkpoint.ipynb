#### Data Cleaning

```python
train = pd.merge(train_transaction, train_identity, on='TransactionID', how='left')
test = pd.merge(test_transaction, test_identity, on='TransactionID', how='left')
```

```python
print(train.shape)
print(test.shape)
```

```python
train_stats = pd.DataFrame(train.isnull().sum()).reset_index()
train_stats.columns = ['cols', 'nulls']
train_stats['percent_nulls'] = 100.0 * train_stats['nulls']/train.shape[0]
train_stats = train_stats.sort_values(by = 'nulls', ascending = False).reset_index(drop = True)
```

```python
train_stats
```

We can see that there are soo many columns with null values more than 70%-75%. Imputing those columns straight away with mean/median for numerical columns or mode for categorical columns would be a bad idea as majority of data is null for these columns. Even if we want to try with bivariate imputation techniques like **KNN imputation** or **Iterative imputation** there are soo many nulls in other columns as well. So the approach we could follow is to select the threshold for null percent of columns to remove if matches the threshold. Remaining columns which do not meet the threshold colud be imputed with measures of central tendency

```python
inc = 0
removed_columns = []
for i in train_stats.values:    
    if(i[2] >= 60.0):
        print(i[0])
        train.drop(columns=i[0],axis=1,inplace=True)
        print("column dropped {} with {} % null values".format(i[0],i[2]))
        inc = inc + 1
        removed_columns.append(i[0])
print("total number of columns dropped {}".format(inc))
```

#### Impute

```python
catcolumns = train.select_dtypes('object').columns

for col in train.columns:
    if train[col].isnull().sum() > 0:
        if col in catcolumns:
            train[col].fillna(train[col].mode()[0],inplace=True)
            print("column {} has been imputed with mode".format(col))
        else:
            
            train[col].fillna(train[col].median(),inplace=True)
            print("column {} has been imputed with mean".format(col))          
```

```python
catcolumns = test.select_dtypes('object').columns

for col in test.columns:
    if test[col].isnull().sum() > 0:
        if col in catcolumns:
            test[col].fillna(test[col].mode()[0],inplace=True)
            print("column {} has been imputed with mode".format(col))
        else:
            
            test[col].fillna(test[col].median(),inplace=True)
            print("column {} has been imputed with mean".format(col))          
```

We need to scale the X_train and X_test using RobustSclaer as it scales the data robust to outliers

**Note** : We were actually supposed to impute and scale by splitting the data first and applying it to train and test seperately in order to avoid **Data Leakage**. Here we are doing that in terms of only scaling. But in real-world production level deployments it is advisable to split and the data into train and test and then apply the scaling/imputing techniques

```python
rs = RobustScaler()
X_train = rs.fit_transform(X_train)
X_val = rs.transform(X_val)
```

```python
X_train = pd.DataFrame(X_train,columns = X.columns)
X_val = pd.DataFrame(X_val,columns = X.columns)
```

```python
X_train.head()
```

**Below is the code for BayesianSearchCV which is the hyperparameter tuning technique which is more accurate than the usual GridSearchCV or RandomizedSearchCV because BayesianSearchCV takes into account of previous hyperparameter learnings to find next best hyperparameters. So there is kind of link between previous and current execution unlike in the case of other two techniques. Unfortunately I couldn't execute it as beacause it doesn't work with any recent sklearn version as it has iid parameter conflict issue. I will try to implement it other project by downgrading sklearn in new virtual envirornment**

```python
# define search space
params = dict()
params['n_estimators'] = (100,120)
params['max_depth'] = (10,15)
params['criterion'] = ['gini','entropy']
params['max_features'] = ['auto', 'sqrt', 'log2']

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
search = BayesSearchCV(estimator=RandomForestClassifier(), search_spaces=params, n_jobs=-1, cv=cv,iid=False)
search.fit(X_train, y_train)

print(search.best_score_)
print(search.best_params_)
```

```python
rfcl = RandomForestClassifier(n_estimators = 100,random_state=1234,max_depth=25,criterion='gini',max_features='sqrt')
rfcl = rfcl.fit(X_train, y_train)
y_predict_rfcl = rfcl.predict(X_val)
print(rfcl.score(X_train , y_train))
print(rfcl.score(X_val, y_val))
print(confusion_matrix(y_val, y_predict_rfcl))
print(classification_report(y_val, y_predict_rfcl))
```

```python
probs = rfcl.predict_proba(X_val)
preds = probs[:,1]
fpr, tpr, threshold = roc_curve(y_val, preds)
roc_auc = auc(fpr, tpr)
metrics
# method to plot
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0,1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

Though the AUC is pretty good we still need to improve the recall for positive class which is 0.43

```python
lr = LogisticRegression(max_iter = 2000)
lr = lr.fit(X_train, y_train)
y_predict_lr = lr.predict(X_val)
print(lr.score(X_train , y_train))
print(lr.score(X_val, y_val))
print(confusion_matrix(y_val, y_predict_lr))
print(classification_report(y_val, y_predict_lr))


probs = lr.predict_proba(X_val)
preds = probs[:,1]
fpr, tpr, threshold = roc_curve(y_val, preds)
roc_auc = auc(fpr, tpr)
metrics
# method to plot
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0,1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
rfcl = RandomForestClassifier(n_estimators = 100,random_state=1234,max_depth=25,criterion='gini',max_features='sqrt',class_weight='balanced')
rfcl = rfcl.fit(X_train, y_train)
y_predict_rfcl = rfcl.predict(X_val)
print(rfcl.score(X_train , y_train))
print(rfcl.score(X_val, y_val))
print(confusion_matrix(y_val, y_predict_rfcl))
print(classification_report(y_val, y_predict_rfcl))

probs = rfcl.predict_proba(X_val)
preds = probs[:,1]
fpr, tpr, threshold = roc_curve(y_val, preds)
roc_auc = auc(fpr, tpr)
metrics
# method to plot
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0,1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

We have slightly improved the recall for minority class at the cost of precision of minority class

**Threshold Moving**
We know that the default threshold value for the classification problem is 0.5. If the probability of the positive class is greater than 0.5 then it is classified as the positive class else negative. Since the class labels are highly imbalanced we need to figure out an optimum threshold to tackle the minority class(positive class) prediction. You can see that the threshold calculated using gmeans technique with the help of roc_auc is **0.034715**(see below). But we can see that after applying the new threshold the precision of the minority class reduced to great extent. It means lot of non-fraud data points are misclassified as fraud data points. According to our business problem we need good recall that is less false negatives. We need less fraud data points to be misclassified as non-fraud because it negatively impacts the business. We can see that the recall we got for minority class is **87%** which is actually pretty good considering the business objective. This is the tradeoff we need to make based on our business problem. We could also try resampling techniques like SMOTE.

```python
#Threshold Moving
from xgboost import XGBClassifier
from numpy import sqrt
from numpy import argmax


xgb_clf = XGBClassifier(random_state=17,n_estimators=175,max_depth=7)
xgb_clf.fit(X_train, y_train)
ypred = xgb_clf.predict_proba(X_val)
ypred = ypred[:, 1]

fpr, tpr, thresholds = roc_curve(y_val, ypred)
gmeans = sqrt(tpr * (1-fpr))
ix = argmax(gmeans)
print('Best Threshold=%f, G-Mean=%.3f' % (thresholds[ix], gmeans[ix]))

plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
plt.plot(fpr, tpr, marker='.', label='XGB')
plt.scatter(fpr[ix], tpr[ix], marker='o', color='black', label='Best')
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
# show the plot
plt.show()
```

```python
thresh = 0.034715
ypred = xgb_clf.predict_proba(X_val)
probs = ypred[:, 1]
ypred = (probs >= thresh).astype('int')

print(accuracy_score(y_val,ypred))
print(confusion_matrix(y_val,ypred))
print(classification_report(y_val,ypred))
```
